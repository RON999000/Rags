# Chat template

```python
chat = [
    {"role":"user","content":"write a python program to build a neural network"}
]
```

```python
prompt = tokenizer.apply_chat_template(chat,tokenize=False,add_generation_prompt=True)
```

```python
input_ids = tokenizer(prompt,add_special_tokens=False,return_tensors="pt").to("cuda")
response = model.generate(**input_ids,max_new_tokens=512)
```

```python
output = tokenizer.decode(response[0])
print(output)
```
